以數學的定義，y、z、w三者的數值都是0.3，且三者相等。
計算機的定義也是如此嗎?

---



```python
x = 0.1
y = x + x + x
z = 3 * x

print('x==0.1? {}'.format(x == 0.1))
print('y==z? {}'.format(y == z))
print('y==0.3? {}'.format(y == 0.3))
print('z==0.3? {}'.format(z == 0.3))
```

```python
w = 0.3

print('y==w? {}'.format(y == w))
print('w==0.3? {}'.format(w == 0.3))
```

```python
print('y   = {:.17f}'.format(y))
print('w   = {:.17f}'.format(w))
print('0.3 = {:.17f}'.format(0.3))
```

```python
print('y  = {:.18f}'.format(y))
print('y == 0.300000000000000045? {}'.format(y == 0.300000000000000045))
```

float.as_integer_ratio(): Return a pair of integers whose ratio is exactly equal to the original float and with a positive denominator. Raises OverflowError on infinities and a ValueError on NaNs.

---



```python
print('integer ratio of x   = {}'.format(x.as_integer_ratio()))
print('integer ratio of 0.1 = {}'.format((0.1).as_integer_ratio()))
print('integer ratio of y   = {}'.format(y.as_integer_ratio()))
print('integer ratio of z   = {}'.format(z.as_integer_ratio()))
print('integer ratio of w   = {}'.format(w.as_integer_ratio()))
print('integer ratio of 0.3 = {}'.format((0.3).as_integer_ratio()))
print('integer ratio of 0.300000000000000045 = {}'.format((0.300000000000000045).as_integer_ratio()))
```
