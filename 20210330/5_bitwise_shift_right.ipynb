The general form of a standard format specifier is:

format_spec ::=  [[fill]align][sign][#][0][width][,][.precision][type]

fill            ::=  \<any character>

align       ::=  "<" | ">" | "=" | "^"

sign        ::=  "+" | "-" | " "

width       ::=  integer

precision   ::=  integer

type        ::=  "b" | "c" | "d" | "e" | "E" | "f" | "F" | "g" | "G" | "n" | "o" | "s" | "x" | "X" | "%"

```python
x = 1
sr_1 = x >> 1
sr_2 = x >> 2

print('x = 0b{:b}'.format(x))
print('shift right 1 of x = {0:#0b} or {0}'.format(sr_1))
print('shift right 2 of x = {0:#0b} or {0}'.format(sr_2))
```

```python
y = 7
sr_1 = y >> 1
sr_2 = y >> 2

print('y = 0b{:b}'.format(y))
print('shift right 1 of y = {0:#0b} or {0}'.format(sr_1))
print('shift right 2 of y = {0:#0b} or {0}'.format(sr_2))
print('floor division of y = {0:#0b} or {0}'.format(y // 2))
```

```python
z = -7
sr_1 = z >> 1
sr_2 = z >> 2

print('z = {:-#0b}'.format(z))
print('shift right 1 of z = {0:-#0b} or {0}'.format(sr_1))
print('shift right 2 of z = {0:-#0b} or {0}'.format(sr_2))
print('floor division of z = {0:#0b} or {0}'.format(z // 2))
```

[高精度計算 (Arbitrary-precision)](https://https://zh.wikipedia.org/wiki/%E9%AB%98%E7%B2%BE%E5%BA%A6%E8%AE%A1%E7%AE%97)

高精度計算是一種程式設計的演算法。由於中央處理器的字長限制，如32位元CPU中一個整數最大只能取值4,294,967,295（=232-1）。因此在進行更大範圍的數值計算中，往往要採取類比手段。通常通過分離字元的方法通過數字陣列進行輸入、通過陣列倒序輸出、通過類比豎式計算進行計算。一般而言，主要類比的是按位元運算，可以用不同的進位制達成不同的目的。

有許多程式庫支援高精度計算，最著名的是GNU多重精度運算庫。另外，Java，Python和Pascal也有原生的高精度運算支援。
