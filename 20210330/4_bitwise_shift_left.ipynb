str.format(*args, **kwargs)


---

replacement_field ::=  "{" [field_name] ["!" conversion] [":" format_spec] "}"

field_name        ::=  arg_name ("." attribute_name | "[" element_index "]")*

arg_name          ::=  [identifier | digit+]

attribute_name    ::=  identifier

element_index     ::=  digit+ | index_string

index_string      ::=  <any source character except "]"> +

conversion        ::=  "r" | "s" | "a"

format_spec       ::=  \<described in the next section>


---




The general form of a standard format specifier is:

format_spec ::=  [[fill]align][sign][#][0][width][,][.precision][type]

fill            ::=  \<any character>

align       ::=  "<" | ">" | "=" | "^"

sign        ::=  "+" | "-" | " "

width       ::=  integer

precision   ::=  integer

type        ::=  "b" | "c" | "d" | "e" | "E" | "f" | "F" | "g" | "G" | "n" | "o" | "s" | "x" | "X" | "%"

```python
x = 1
sl_1 = x << 1
sl_2 = x << 2
sl_3 = x << 3

print('x = 0b{:b}'.format(x))
print('shift left 1 of x = ob{:b} or {}'.format(sl_1, sl_1))
print('shift left 2 of x = ob{0:b} or {0}'.format(sl_2))
print('shift left 3 of x = ob{0:b} or {0}'.format(sl_3))
```

```python
import time

start = time.time()
for i in range((2 ** 31) + 0, (2 ** 31) + 20000000):
    x = i * 2

end = time.time()
print('Spent times of multiplier in large number = {}'.format(end - start))
```
