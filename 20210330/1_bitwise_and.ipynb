bin(x): built-in function. Convert an integer number to a binary string prefixed with “0b”. 

---

冒號之後的格式語法
 
The general form of a standard format specifier is:

format_spec ::=  [[fill]align][sign][#][0][width][,][.precision][type]

fill        ::=  \<any character>

align       ::=  "<" | ">" | "=" | "^"

sign        ::=  "+" | "-" | " "

width       ::=  integer

precision   ::=  integer

type        ::=  "b" | "c" | "d" | "e" | "E" | "f" | "F" | "g" | "G" | "n" | "o" | "s" | "x" | "X" | "%"

---

```python
x = 255
bin_x = bin(x)

print('x = {}'.format(bin_x))
print('x = {:b}'.format(x))
print('type of bin_x = {}'.format(type(bin_x)))
print('type of x = {}'.format(type(x)))
```

```python
x = 170
y = 85
z = 15
x_and_y = x & y
x_and_z = x & z

print('x = {:08b}'.format(x))
print('y = {:08b}'.format(y))
print('z = {:08b}'.format(z))
print('x & y = {:08b}'.format(x_and_y))
print('x & z = {:08b}'.format(x_and_z))
```
