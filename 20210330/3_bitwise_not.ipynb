The general form of a standard format specifier is:

format_spec ::=  [[fill]align][sign][#][0][width][,][.precision][type]

fill        ::=  \<any character>

align       ::=  "<" | ">" | "=" | "^"

sign        ::=  "+" | "-" | " "

width       ::=  integer

precision   ::=  integer

type        ::=  "b" | "c" | "d" | "e" | "E" | "f" | "F" | "g" | "G" | "n" | "o" | "s" | "x" | "X" | "%"


---



```python
x = 170
y = -171
not_x = ~x
not_y = ~y

print('    x = {:>9b}'.format(x))
print('not_x = {:>9b}'.format(not_x))
print('not_x = {:>9}'.format(not_x))
print('    y = {:>9b}'.format(y))
print('not_y = {:>9b}'.format(not_y))
print('not_y = {:>9}'.format(not_y))
```

int.to_bytes(length, byteorder, *, signed=False): Return an array of bytes representing an integer.

---

int.bit_length(): Return the number of bits necessary to represent an integer in binary, excluding the sign and leading zeros.

---



[一補數](https://zh.wikipedia.org/wiki/%E4%B8%80%E8%A3%9C%E6%95%B8)

將二進位數每個數字反轉，得到的數即為原二進位的一補數（英語：ones' complement）。若某一位為0，則使其變為1，反之亦然。

---



[二補數](https://zh.wikipedia.org/wiki/%E4%BA%8C%E8%A3%9C%E6%95%B8)

二補數（英語：2's complement）是一種用二進位表示有號數的方法，也是一種將數字的正負號變號的方式，常在電腦科學中使用。二補數以有符號位元的二進位數定義。

正數和0的二補數就是該數字本身。負數的二補數則是將其對應正數按位元取反再加1。


---



```python
signed = True
bin_y = bin(y.to_bytes((y.bit_length() + signed + 7) // 8, 'little', signed=signed)[0])
bin_not_y = bin(not_y.to_bytes((not_y.bit_length() + signed + 7) // 8, 'little', signed=signed)[0])

print('    y = {}'.format(bin_y))
print('not_y = {}'.format(bin_not_y))
```
