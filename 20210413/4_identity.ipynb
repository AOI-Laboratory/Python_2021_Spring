**id(object)**

Return the “identity” of an object.

**短字串**

```python
s1 = 'short'
s2 = 'short'

if s1 is s2:
    print('s1 and s2 are the same object')
else:
    print('s1 and s2 are \"different\" objects')

print('ID of s1 = {}, ID of s2 = {}'.format(id(s1), id(s1)))

if s1 == s2:
    print('s1 and s2 have the same value')
else:
    print('s1 and s2 have \"different\" value')
```

**長字串**

```python
s3 = 'This is a longer string'
s4 = 'This is a longer string'

if s3 is s4:
    print('s3 and s4 are the same object')
else:
    print('s3 and s4 are \"different\" objects')

print('ID of s3 = {}, ID of s4 = {}'.format(id(s3), id(s4)))

if s3 == s4:
    print('s3 and s4 have the same value')
else:
    print('s3 and s4 have \"different\" value')
```

**Interning**

最佳化的一種技術，重複使用或快取(chache)相同內容的物件，可以減少記憶體的使用量和物件初始化的時間。

必須是***mutable***物件才能使用interning最佳化技術。


**Integer interning規則**

整數值介於***-5***到***256***之間的整數物件才會使用interning最佳化技術。

**String interning規則**

*   compile time
*   長度小於20字元
*   只能是ASCII字母、數字和底線
