**[ ]**

slicing operator

index是正整數時，一律從0開始

```python
alpha= 'abcdefghijklmnopqrstuvwxyz'

print('alpha[0] = {}'.format(alpha[0]))
print('alpha[1] = {}'.format(alpha[1]))
print('alpha[2] = {}'.format(alpha[2]))
```

index也可以是負數，從-1開始

```python
print('alpha[-1]] = {}'.format(alpha[-1]))
print('alpha[-2]] = {}'.format(alpha[-2]))
print('alpha[-3]] = {}'.format(alpha[-3]))
```

index超出字串長度會引發runtime錯誤

alpha的index從0到25總共26個

```python
print('alpha[26]] = {}'.format(alpha[26]))
```

alpha的index從-1到-26總共26個

```python
print('alpha[-27]] = {}'.format(alpha[-27]))
```

index還可以是一個範圍，正常範圍是從小到大。
index從大到小不會出錯，會回傳空字串。

**string\[x:y\]**  index從 x 到 (y - 1)

**string\[:y\]**  index從 0 到 (y - 1)

**string\[x:-y\]**  index從 x 到 (len(string) - y - 1)

**string\[:-y\]**  index從 0 到 (len(string) - y - 1)


**string\[x:\]**  index從 x 到 len(string)

**string\[-x:\]**  index從 (len(string) - x) 到 len(string)

**string\[::-1\]**  反轉字串


```python
print('alpha[0:25] = {}'.format(alpha[0:25]))
print('alpha[:25] = {}'.format(alpha[:25]))
print('alpha[0:-1]] = {}'.format(alpha[0:-1]))
print('alpha[:-1]] = {}'.format(alpha[:-1]))
print('alpha[1:] = {}'.format(alpha[1:]))
print('alpha[-25:] = {}'.format(alpha[-25:]))
print('alpha[::-1] = {}'.format(alpha[::-1]))
```

**comparision operators**


```python
if s2 == '12345':
    print("s2 is equal to '12345'")
else:
    print("s2 and '12345' are different")

if s2 != '01234':
    print("s2 and '01234' are different")
else:
    print("s2 is equal to '01234'")
```

**membership operators**

```python
if 'operator' in s1:
    print("s1 includes 'operator'")
else:
    print("s1 does not include 'operator'")

if 'operators' not in s1:
    print("s1 does not include 'operators'")
else:
    print("includes 'operators'")
```
