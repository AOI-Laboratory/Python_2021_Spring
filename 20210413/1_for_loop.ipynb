**class range(stop)**

**class range(start, stop[, step])**

Rather than being a function, range is actually an immutable sequence type, as documented in Ranges and Sequence Types — list, tuple, range.

**示範for ... else ...**

```python
for i in range(5):
    print('i = {}'.format(i))
else:
    print('for-loop finished!')
```

**示範continue和else**

```python
for i in range(5):
    if i == 3:
        continue
    print('i = {}'.format(i))
else:
    print('for-loop finished!')
```

**示範break和else**

```python
for i in range(5):
    if i == 3:
        break
    print('i = {}'.format(i))
else:
    print('for-loop finished!')
```
