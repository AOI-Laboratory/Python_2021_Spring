**dir([object])**

Without arguments, return the list of names in the current local scope. With an argument, attempt to return a list of valid attributes for that object.

```python
print('String supports {}'.format(dir('')))
```

```python
print('list supports {}'.format(dir(list())))
```

```python
print('range supports {}'.format(dir(range(0))))
```

**示範字串的+運算子(concatenate operator)**

```python
import string

letters = string.ascii_letters + string.digits + string.punctuation
print('letters = {}'.format(letters))
```

**str.join(iterable)**

Return a string which is the concatenation of the strings in iterable. A TypeError will be raised if there are any non-string values in iterable, including bytes objects. The separator between elements is the string providing this method.

**random.choice(seq)**

Return a random element from the non-empty sequence seq. If seq is empty, raises IndexError.

**從letters隨機選取10個字元**

```python
import random

text = ''.join(random.choice(letters) for i in range(10))

print('text = {}'.format(text))
```

**str.isalpha()**

Return True if all characters in the string are alphabetic and there is at least one character, False otherwise.

**示範字串(string)的iterable** : 搜尋不是字母(alphabet)的字元(character)

方法一: 使用字串的成員函式(member function) isalpha

```python
for c in text:
    if not c.isalpha():
        print('{} is not an alphabet'.format(c))
```

**示範字串(string)的iterable**

方法二: 使用字串的[]運算子(slicing operator)

```python
for idx in range(len(text)):
    c = text[idx]
    if not c.isalpha():
        print('Character position {} : {} is not an alphabet'.format(idx,c))
```

**示範字串(string)的iterable**

方法三: 使用字串的not in運算子(membership operator)

```python
for c in text:
    if c not in string.ascii_letters:
        print('{} is not an alphabet'.format(c))
```

**示範串列(list)的iterable**

方法一: 使用串列的==運算子(comparision operator)

```python
members = dir(list())
for attr in members:
    if attr == '__iter__' or attr == '__getitem__' or attr == '__next__':
        print('{} is one of list member functions'.format(attr))
```

**示範串列(list)的iterable**

方法二: 使用串列的in運算子(membership operator)**

```python
targets = ['__iter__', '__getitem__', '__next__']
for function in targets:
    if function in members:
        print('{} is one of list member functions'.format(function))
```
