# Caller引數(argument)和function參數(parameter)匹配規則


## Positional argument

引數和參數依照位置，一個接著一個依序匹配

Positional_Argument.png

```python
def func1(a, b, c):
    print('a is {}, b is {} and c is {}'.format(a, b, c))
```

```python
func1(1, 2, 3)
```

```python
func1(2, 3, 1)
```

positional argument不能省略任何一個引數

```python
func1(1, 2)
```

## Keyword Argument
呼叫函數時，使用參數的名稱設定引數，而不是參數的位置。

```python
func1(a=1, b=2, c=3)
```

引數的位置沒有跟參數的位置匹配也沒關係

```python
func1(b=2, c=3, a=1)
```

# 必需要的引數(required argument)和選擇性引數(optional argument)

## Optional argument
optional argument是有**預設值**的

非optional argument是required argument，也就是**沒有**預設值的




```python
def func2(x=4, y=5, z=6):
    print('x is {}, y is {} and z is {}'.format(x, y, z))
```

```python
func2()
```

使用位置匹配引數和參數

```python
func2(10)
```

```python
func2(10, 20)
```

使用參數的名稱設定引數，引數和參數的位置可以不需要匹配

```python
func2(z=20, x=10)
```

# 混合required arguments和optional arguments

規則是required arguments在前，optional arguments在後

換句話說，沒有預設值的放前面，有預設值的放後面

```python
def func3(a, b, c, x=4, y=5, z=6):
    print('Required: a is {}, b is {} and c is {}'.format(a, b, c))
    print('Optional: x is {}, y is {} and z is {}'.format(x, y, z))
```

```python
func3(1, 2, 3)
```

```python
func3(1, 2, 3, z=30)
```

```python
func3(z=30, b=2, c=3, a=1)
```

*沒有預設值的放前面，有預設值的放後面*

```python
func3(z=30, 1, 2, 3)
```

# Arbitrary argument \*args and **kwargs

函數可以接收任意個引數

|          | Data type | Argument type                 |
|----------|-----------|-------------------------------|
| *args    | tuple     | unmatched positional argument |
| **kwargs | dict      | unmatched keyword argument    |

```python
def accumulator(*args):
  sum = 0
  for num in args:
    sum += num
  
  print('sum = {}'.format(sum))
```

```python
accumulator(1, 2, 3, 4, 5, 6, 7, 8, 9)
```

```python
def func4(**kwargs):
  for key in sorted(kwargs):
    print('{} = {}'.format(key, kwargs[key]))
```

```python
func4(x=4, c=3, y=5, z=6, a=1, b=2)
```

### 混合所有不同類型的引數

順序: positional arguments --> *args --> keyword arguments --> **kwargs

```python
def func5(a, *args, x=4, **kwargs):
  print('positional argument a is {}'.format(a))
  print('args = {}'.format(args))
  print('optional argument x is {}'.format(x))
  print('kwargs = {}'.format(kwargs))
```

```python
func5(1, 2, 3, x=4, y=5, z=6)
```

```python
func5(1, 2, 3, y=5, z=6)
```

```python
func5(1, 2, 3, 4, y=5, z=6)
```

**順序錯誤**

```python
def func6(a, **kwargs, *args, x=4):
  print('positional argument a is {}'.format(a))
  print('args = {}'.format(args))
  print('optional argument x is {}'.format(x))
  print('kwargs = {}'.format(kwargs))
```
