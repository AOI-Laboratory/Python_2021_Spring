**空函式的宣告與呼叫**


```python
def func1():
    pass

def func2():
  return

func1()
func2()
```

**無參數函式的宣告與呼叫**

```python
def hi():
  print('Hi!!')

hi()
hi()
hi()
```

**參數函式的宣告與呼叫**

```python
def hi(name):
    print('Hi, {}!!'.format(name))

hi('Michael')
hi('James')
hi('Kobe')
```

**參數的預設值**

```python
def hi(name='everyone'):
    print('Hi, {}!!'.format(name))

hi('Michael')
hi('James')
hi('Kobe')
hi()
```

**函式的回傳值**

```python
def plus_one(int_var):
  return int_var + 1

for i in range(10):
  print(plus_one(i), end=' ')
```
